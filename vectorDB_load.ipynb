```python
#from langchain.embeddings import HuggingFaceEmbeddings
#from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_huggingface import HuggingFaceEmbeddings

#from langchain.vectorstores import FAISS
from langchain_community.vectorstores import FAISS

#from langchain.schema import Document
from langchain_core.documents import Document  # ‚úÖ ÏµúÏã† ÏúÑÏπò
import json
```

```python
#Ï†ÄÏû•Îêú Ïä§ÌÅ¨ÎûòÌïë Îç∞Ïù¥ÌÑ∞ Î°úÎìú
#with open("results_all.json", "r", encoding = "utf-8") as f:
with open("unique_meta.json", "r", encoding = "utf-8") as f:
    loaded_docs = json.load(f)
```

```python
# ‚úÖ ÏûÑÎ≤†Îî© Î™®Îç∏ ÏÑ§Ï†ï (GPU ÏÑúÎ≤ÑÏóêÏÑú Ïã§Ìñâ Í∞ÄÎä•)
#embedder = HuggingFaceEmbeddings(model_name="jhgan/ko-sbert-nli")

embedder = HuggingFaceEmbeddings(model_name="intfloat/multilingual-e5-small")
```

```python
# ‚úÖ Ï†ÄÏû•Îêú DB Î°úÎìú
loaded_db = FAISS.load_local("faiss_langchain_db_meta_v2", embeddings=embedder, allow_dangerous_deserialization=True)
```

```python

#query = "What is LangChain?"
query = "How to use the LangChain library?"
retrieved_docs = loaded_db.similarity_search_with_score(query, k=30)

for i, (doc, score) in enumerate(retrieved_docs, 1):
    # if Î¨∏ÏÑú ÌÉÄÏûÖÏù¥ 'text'Ïù∏ Í≤ΩÏö∞ÏóêÎßå Ï∂úÎ†•
    # ÏÉÅÌô©Ïóê Îî∞ÎùºÏÑú Î©îÌÉÄ Îç∞Ïù¥ÌÑ∞ Ï°∞Í±¥ Ï∂îÍ∞Ä Í∞ÄÎä•, Ï†ÑÎ∂Ä Îã§ Ï∂úÎ†•ÌïòÎ†§Î©¥ if Î¨∏ Ï†úÍ±∞
    if doc.metadata['type'] == 'text':
        print(f"--- Retrieved Document {i} ---")
        print(f"Title: {doc.metadata['title']}")
        print(f"Type: {doc.metadata['type']}")
        print(f"Side Link: {doc.metadata['side_link']}")
        print(f"Head Menu Name: {doc.metadata['head_menu_name']}")
        print(f"Side Menu Name: {doc.metadata['side_menu_name']}")
        print(f"Content: {doc.page_content[:200]}...")  # ÏïûÎ∂ÄÎ∂Ñ 200ÏûêÎßå Ï∂úÎ†•
        print(f"Score: {score}")
        print()
```
